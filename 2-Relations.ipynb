```python
# =-=-=-=-=-=-=-=-=-=-=
# Imports
# =-=-=-=-=-=-=-=-=-=-= 
import re, nltk
import pandas as pd
from stop_words import get_stop_words
stopwords = get_stop_words('en')

# =-=-=-=-=-=-=-=-=-=-=
# Ingest and listify the text
# =-=-=-=-=-=-=-=-=-=-=
mdg = open('texts/most_dangerous_game.txt', 'r').read()
words = re.sub("[^'a-zA-Z']"," ", mdg).lower().split()
```

## Word Distributions

```python
text = nltk.Text(words)
```

```python
text.dispersion_plot(["rainsford", "general", "said", "hunting", "night", "eyes", "zaroff", "heard", "hunt", "man"])
```
